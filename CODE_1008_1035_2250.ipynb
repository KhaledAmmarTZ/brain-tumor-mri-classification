Brain Tumor MRI Classification using Custom CNN (TensorFlow)

Dataset: Binary Classification (Tumor / No Tumor)

```python


import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import seaborn as sns
import os, zip

DATASET_ROOT = "/content/Dataset"

train_dir = os.path.join(DATASET_ROOT, "Training")
test_dir  = os.path.join(DATASET_ROOT, "Testing")

print("Training folder:", train_dir)
print("Testing folder:", test_dir)


# Image Preprocessing with augmentation

img_height, img_width = 150, 150
batch_size = 32

train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

test_gen = ImageDataGenerator(rescale=1./255)

train_data = train_gen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=True
)

test_data = test_gen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)

print("Classes:", train_data.class_indices)


# Build Custom CNN

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.5),

    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# Train the Model

epochs = 20

history = model.fit(
    train_data,
    validation_data=test_data,
    epochs=epochs
)

# Plot Accuracy & Loss

plt.figure(figsize=(12, 5))

# Accuracy Plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc', marker='o')
plt.plot(history.history['val_accuracy'], label='Val Acc', marker='o')
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()

# Loss Plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss', marker='o')
plt.plot(history.history['val_loss'], label='Val Loss', marker='o')
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()


# Evaluation

loss, acc = model.evaluate(test_data)
print("\n==============================")
print(f" Final Test Accuracy: {acc*100:.2f}%")
print(f" Final Test Loss: {loss:.4f}")
print("==============================")


# Confusion Matrix

y_true = test_data.classes
y_pred_prob = model.predict(test_data)
y_pred = (y_pred_prob > 0.5).astype(int).flatten()

cm = confusion_matrix(y_true, y_pred)
labels = ["No Tumor", "Tumor"]

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=labels, yticklabels=labels)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=labels))


# Save Model

model.save("/content/brain_tumor_custom_cnn.h5")
print("\nModel saved successfully!")
```


E D A for Brain Tumor MRI Dataset (Training + Testing)

Two Classes: Tumor / No_Tumor

```python


import matplotlib.pyplot as plt
import random
from pathlib import Path
import os

DATASET_ROOT = "/content/drive/MyDrive/Dataset"
IMG_EXT = ('.png', '.jpg', '.jpeg')

train_dir = Path(DATASET_ROOT) / "Training"
test_dir  = Path(DATASET_ROOT) / "Testing"

print("Training path:", train_dir)
print("Testing path:", test_dir)

def count_images(root):
    counts = {}
    for cls in sorted(root.iterdir()):
        if cls.is_dir():
            imgs = [f for f in cls.iterdir() if f.suffix.lower() in IMG_EXT]
            counts[cls.name] = len(imgs)
    return counts

def get_samples(root, n_samples=2):
    samples = {}
    for cls in sorted(root.iterdir()):
        if cls.is_dir():
            imgs = [f for f in cls.iterdir() if f.suffix.lower() in IMG_EXT]
            if len(imgs) > 0:
                samples[cls.name] = random.sample(imgs, min(len(imgs), n_samples))
    return samples

#  Count Training & Testing images
train_counts = count_images(train_dir)
test_counts  = count_images(test_dir)

print("\nTraining counts:", train_counts)
print("Testing counts:", test_counts)

#  Plot Class Distribution for Training & Testing
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.bar(train_counts.keys(), train_counts.values(), color=['#4CAF50','#E91E63'])
plt.title("Training Set Distribution")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.grid(axis='y')

plt.subplot(1,2,2)
plt.bar(test_counts.keys(), test_counts.values(), color=['#4CAF50','#E91E63'])
plt.title("Testing Set Distribution")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.grid(axis='y')

plt.tight_layout()
plt.show()

# Images Sample per Class (Training set)
samples = get_samples(train_dir, n_samples=2)

plt.figure(figsize=(12,6))

idx = 1
for cls_name, imgs in samples.items():
    for img_path in imgs:
        plt.subplot(len(samples), 2, idx)
        img = plt.imread(img_path)
        plt.imshow(img, cmap='gray')
        plt.title(f"{cls_name}")
        plt.axis("off")
        idx += 1

plt.tight_layout()
plt.show()
```

Pre-trained model(MobileNetV2)

Binary Classification: Tumor vs No Tumor

```python


import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
from pathlib import Path

DATA_ROOT = "/content/drive/MyDrive/Dataset"
train_dir = Path(DATA_ROOT) / "Training"
test_dir  = Path(DATA_ROOT) / "Testing"

IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32

# Preprocessing & Augmentation

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=True
)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

print("\nClass mapping:", train_gen.class_indices)

# Load Pre-trained Model -MobileNetV2-

base_model = tf.keras.applications.MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)
)

base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.4)(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=0.0001),
    metrics=['accuracy']
)

model.summary()

EPOCHS = 15

history = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=EPOCHS
)

# Plot Accuracy & Loss

plt.figure(figsize=(12,5))

# Accuracy Plot
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], marker='o', label='Train Acc')
plt.plot(history.history['val_accuracy'], marker='o', label='Val Acc')
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()

# Loss Plot
plt.subplot(1,2,2)
plt.plot(history.history['loss'], marker='o', label='Train Loss')
plt.plot(history.history['val_loss'], marker='o', label='Val Loss')
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Evaluate on Test Set

loss, acc = model.evaluate(test_gen)
print(f"\n✅ Test Accuracy: {acc*100:.2f}%")
print(f"🏁 Test Loss: {loss:.4f}")


y_true = test_gen.classes
y_pred_prob = model.predict(test_gen)
y_pred = (y_pred_prob > 0.5).astype(int).flatten()

cm = confusion_matrix(y_true, y_pred)
labels = ["No Tumor", "Tumor"]

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=labels, yticklabels=labels)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=labels))

model.save("/content/brain_tumor_mobilenetv2.h5")
print("Model saved successfully!")
```

Fine-tuned VGG16 for Brain Tumor MRI Classification

```python

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

DATA_ROOT = "/content/drive/MyDrive/Dataset"
train_dir = DATA_ROOT + "/Training"
test_dir  = DATA_ROOT + "/Testing"

IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32

# Preprocessing
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.20,
    horizontal_flip=True
).flow_from_directory(
    train_dir, target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE, class_mode='binary'
)

test_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    test_dir, target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE, class_mode='binary', shuffle=False
)

# Load VGG16
base = tf.keras.applications.VGG16(include_top=False, weights='imagenet',
                                   input_shape=(IMG_HEIGHT, IMG_WIDTH, 3))
base.trainable = False

x = Flatten()(base.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(base.input, output)
model.compile(optimizer=Adam(0.0005), loss='binary_crossentropy', metrics=['accuracy'])

history1 = model.fit(train_gen, validation_data=test_gen, epochs=10)

# Fine-tune last 4 blocks
for layer in base.layers[-8:]:
    layer.trainable = True

model.compile(optimizer=Adam(1e-5), loss='binary_crossentropy', metrics=['accuracy'])
history2 = model.fit(train_gen, validation_data=test_gen, epochs=10)
model.summary()

# Metrics
y_true = test_gen.classes
y_pred = (model.predict(test_gen) > 0.5).astype(int)

print(classification_report(y_true, y_pred))

cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True)
plt.show()

model.save("/content/vgg16_finetuned.h5")
```

Fine-tuned DenseNet121 for Brain Tumor MRI Classification

```python

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
from pathlib import Path

# Dataset Path

DATA_ROOT = "/content/drive/MyDrive/Dataset"
train_dir = Path(DATA_ROOT) / "Training"
test_dir  = Path(DATA_ROOT) / "Testing"

IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    zoom_range=0.25,
    width_shift_range=0.15,
    height_shift_range=0.15,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=True
)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=False
)

print("\nClass mapping:", train_gen.class_indices)

# Load Pretrained DENSENET121

base = DenseNet121(
    weights="imagenet",
    include_top=False,
    input_shape=(224,224,3)
)

base.trainable = False

x = GlobalAveragePooling2D()(base.output)
x = Dense(256, activation="relu")(x)
x = Dropout(0.4)(x)
out = Dense(1, activation="sigmoid")(x)

model = Model(inputs=base.input, outputs=out)

model.compile(
    optimizer=Adam(1e-4),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# Stage 1 Training only classifier

EPOCHS_STAGE1 = 8

history1 = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=EPOCHS_STAGE1
)

# DENSENET121 Last 40 Layers

for layer in base.layers[-40:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(1e-5),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

EPOCHS_STAGE2 = 10

history2 = model.fit(
    train_gen,
    validation_data=test_gen,
    epochs=EPOCHS_STAGE2
)

# Conbination of history

history = {
    "accuracy": history1.history["accuracy"] + history2.history["accuracy"],
    "val_accuracy": history1.history["val_accuracy"] + history2.history["val_accuracy"],
    "loss": history1.history["loss"] + history2.history["loss"],
    "val_loss": history1.history["val_loss"] + history2.history["val_loss"]
}

plt.figure(figsize=(12,5))

# Accuracy
plt.subplot(1,2,1)
plt.plot(history["accuracy"], marker="o", label="Train Acc")
plt.plot(history["val_accuracy"], marker="o", label="Val Acc")
plt.axvline(EPOCHS_STAGE1, color="red", linestyle="--", label="Fine-tuning Start")
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()

# Loss
plt.subplot(1,2,2)
plt.plot(history["loss"], marker="o", label="Train Loss")
plt.plot(history["val_loss"], marker="o", label="Val Loss")
plt.axvline(EPOCHS_STAGE1, color="red", linestyle="--", label="Fine-tuning Start")
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

loss, acc = model.evaluate(test_gen)
print(f"\n🔥 FINAL Fine-tuned Test Accuracy: {acc*100:.2f}%")
print(f"📉 Final Test Loss: {loss:.4f}")

y_true = test_gen.classes
y_pred_prob = model.predict(test_gen)
y_pred = (y_pred_prob > 0.5).astype(int).flatten()

cm = confusion_matrix(y_true, y_pred)
labels = ["No Tumor", "Tumor"]

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=labels, yticklabels=labels)
plt.title("Confusion Matrix (Fine-tuned DenseNet121)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred, target_names=labels))

model.save("/content/fine_tuned_densenet121_brain_tumor.h5")
print("\n Fine-tuned DenseNet121 model saved successfully!")
```
